```python
import pygplates
import numpy as np
import paleogeography as pg
import glob, re
import os


%matplotlib inline
%load_ext autoreload
%autoreload 2


basedir = '../paleogeography/Paleogeography_Matthews2016_410-2Ma_Shapefiles/'


# make a sorted list of the (midpoint) times for paleogeography polygons
tmp = glob.glob('%s/*/' % basedir)

paleogeography_timeslice_list = []
for tm in tmp:
    paleogeography_timeslice_list.append(float(re.findall(r'\d+Ma+',tm)[1][:-2]))

paleogeography_timeslice_list.sort()
paleogeography_timeslice_list = np.array(paleogeography_timeslice_list)

print paleogeography_timeslice_list

# Note that the order here is important (for the order they are rendered in GPlates)
env_list = ['dm','sm','lm','m','i']

all_pg_features = []

for paleogeography_timeslice in paleogeography_timeslice_list:

    pg_dir = '%s/PresentDay_Paleogeog_Matthews2016_%dMa/' % (basedir,paleogeography_timeslice)

    for env in env_list:
        pg_features = pg.load_paleogeography(pg_dir,[env],env_field='Layer')

        all_pg_features.extend(pg_features)


pygplates.FeatureCollection(all_pg_features).write('Palogeography_0_410Ma.gpmlz')

```
